```python
missing_values = df.isnull().sum()

missing_values
```

```python
# Convert date columns to datetime format
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')
df['DueDate'] = pd.to_datetime(df['DueDate'], errors='coerce')
df['ShipDate'] = pd.to_datetime(df['ShipDate'], errors='coerce')
```

```python
df
```

```python
unique_values_per_column = df.nunique()
unique_values_per_column.reset_index(name='Unique Values').rename(columns={'index': 'Column Name'})
```

```python
duplicates = df[df.duplicated()]
num_duplicates = duplicates.shape[0]
num_duplicates
```

```python
df = df.drop(columns=['DueDate', 'ShipDate'])
```

```python
df
```

```python
df['Profit_Per_Unit'] = df['UnitPrice'] - df['Unit_Cost'] - df['Unit_Freight_Cost'] - df['UnitPriceDiscount']*df['UnitPrice']
df['Total_Profit'] = df['Profit_Per_Unit'] * df['OrderQty']
```

```python
df
```

```python
import matplotlib.pyplot as plt

# Aggregating data by month
monthly_sales = df.groupby(df['OrderDate'].dt.to_period('M')).agg({
    'Total_Profit': 'sum',
    'OrderQty': 'sum',
    'UnitPrice': 'mean',
    'Profit_Per_Unit': 'mean'
}).reset_index()

# Converting period to datetime for plotting
monthly_sales['OrderDate'] = monthly_sales['OrderDate'].dt.to_timestamp()

# Plotting Total Profit Over Time
plt.figure(figsize=(14, 7))
plt.plot(monthly_sales['OrderDate'], monthly_sales['Total_Profit'], marker='o', linestyle='-')
plt.title('Total Profit Over Time')
plt.xlabel('Date')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

# Plotting Average Profit Per Unit Over Time
plt.figure(figsize=(14, 7))
plt.plot(monthly_sales['OrderDate'], monthly_sales['Profit_Per_Unit'], marker='o', linestyle='-')
plt.title('Average Profit Per Unit Over Time')
plt.xlabel('Date')
plt.ylabel('Profit Per Unit')
plt.grid(True)
plt.show()
```

```python

# Analyzing the product categories
category_performance = df.groupby('Product_Category').agg({
    'Total_Profit': 'sum',
    'OrderQty': 'sum',
    'Profit_Per_Unit': 'mean'
}).reset_index()

# Visualize performance by product category
plt.figure(figsize=(14, 7))
plt.bar(category_performance['Product_Category'], category_performance['Total_Profit'])
plt.title('Total Profit by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

# Analyzing the sales regions
region_performance = df.groupby('Sales_Region').agg({
    'Total_Profit': 'sum',
    'OrderQty': 'sum',
    'Profit_Per_Unit': 'mean'
}).reset_index()

# Visualize performance by sales region
plt.figure(figsize=(14, 7))
plt.bar(region_performance['Sales_Region'], region_performance['Total_Profit'])
plt.title('Total Profit by Sales Region')
plt.xlabel('Sales Region')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
# Customer analysis
customer_performance = df.groupby('Customer_Code').agg({
    'Total_Profit': 'sum',
    'OrderQty': 'sum',
    'Profit_Per_Unit': 'mean'
}).reset_index()

# Top 10 customers by total profit
top_customers = customer_performance.nlargest(10, 'Total_Profit')

# Visualize top 10 customers
plt.figure(figsize=(14, 7))
plt.bar(top_customers['Customer_Code'], top_customers['Total_Profit'])
plt.title('Top 10 Customers by Total Profit')
plt.xlabel('Customer Code')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt


def label_encode_categorical(df):
    le = LabelEncoder()
    categorical_columns = df.select_dtypes(include=['object', 'category']).columns
    for col in categorical_columns:
        df[col] = le.fit_transform(df[col])
    return df

transformed_data = label_encode_categorical(df)


correlation_matrix = transformed_data.corr()


def list_important_correlations(correlation_matrix, threshold=0.5):
    important_correlations = []

    for col in correlation_matrix.columns:
        for row in correlation_matrix.index:
            if row != col and abs(correlation_matrix.loc[row, col]) > threshold:
                important_correlations.append((row, col, correlation_matrix.loc[row, col]))

    return important_correlations

important_correlations = list_important_correlations(correlation_matrix)
for item in important_correlations:
    print(f"Correlation between {item[0]} and {item[1]}: {item[2]:.2f}")


plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix with Label Encoded Categorical Variables')
plt.show()
```
